```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
import pylab as pl
rus=pd.read_csv("russian_alcohol_consumption.csv")
```

```python
pl.figure(figsize=(10,15))
sn.barplot(x=rus["year"],y=rus["beer"],color="red")
sn.barplot(x=rus["year"],y=rus["vodka"])
sn.barplot(x=rus["year"],y=rus["wine"])
sn.barplot(x=rus["year"],y=rus["champagne"])
```

```python
pl.figure(figsize=(10,15))#estetico
plt.bar(rus["year"],rus["beer"],color="red")
plt.bar(rus["year"],rus["vodka"])
plt.bar(rus["year"],rus["wine"])
plt.bar(rus["year"],rus["champagne"])
```

```python
sn.scatterplot(data=rus)#data o obrigatorio
```
